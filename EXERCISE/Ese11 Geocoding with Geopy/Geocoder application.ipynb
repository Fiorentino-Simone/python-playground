```python
import pandas
```

```python
df = pandas.read_csv("supermarkets.csv")
df
```

```python
#il geocoder richiede: La via, la città, lo stato con il codice ZIP (cioè il codice postale)
#quindi creiamo la colonna con tutte le informazioni presenti
df["Address"] = df["Address"] + "," + df["City"] + "," + df["State"] + "," + df["Country"]
```

```python
df
```

```python
#al posto di fare un for loop, ci sono dei metodi pandas che prendono tutti i valori di una colonna e questo si chiama apply
df["Coordinates"] = df["Address"].apply(nom.geocode)
df
```

```python
#creiamo altre due colonne in cui inseriamo i due valori delle lat e long
#utilizziamo sempre il metodo apply per prendere tutti i valori contenuti in Coordinates
#ricordo quei valori SONO PROPRIETA' DI UN OGGETTO quindi:
df["Latitude"] = df["Coordinates"].apply(lambda x : x.latitude if x != None else None)
df["Longitude"] = df["Coordinates"].apply(lambda x : x.longitude if x != None else None)
#la lambda sarebbe come il la funzione anonima del JS, praticamente andiamo ad inserire in una variabile temporanea il valore della colonna
#andiamo poi a prendere la latitudine solamente se il valore della x è diverso da None senno si inserisce None nella variabile
#esempio ulteriore del lambda:
#a = foo(lambda x: x*2, 5)
#output >>> 10
df
```
